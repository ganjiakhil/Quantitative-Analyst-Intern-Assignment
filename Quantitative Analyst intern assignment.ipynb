```python
import requests
import pandas as pd
#from datetime import datetime
```

```python
def run_query(q):
    request = requests.post('https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2', json={'query': query})
    if request.status_code == 200:
        return request.json()
    else:
        raise Exception('Query failed. return code is {}, {}'.format(request.status_code, query))
```

```python
query = """
{
  pairs(first: 1000, orderBy: createdAtTimestamp, orderDirection: desc) {
    id
    volumeToken0
    volumeToken1
    volumeUSD
    createdAtTimestamp
    token0{
      symbol
      totalLiquidity
    }
    token1{
      symbol
      totalLiquidity
    }
  }
}
"""

result = run_query(query)
```

```python
result
```

```python
data = result['data']['pairs']
```

```python
pairCount = len(result['data']['pairs'])
```

```python
timestamp = []
id = []
token0 = []
volumetoken0 = []
token1 = []
volumetoken1 = []
volumeUSD = []
```

```python
for pair in range(pairCount):
    timestamp.append(data[pair]['createdAtTimestamp'])
    id.append(data[pair]['id'])
    token0.append(data[pair]['token0']['symbol'])
    volumetoken0.append(data[pair]['volumeToken0'])
    token1.append(data[pair]['token1']['symbol'])
    volumetoken1.append(data[pair]['volumeToken1'])
    volumeUSD.append(data[pair]['volumeUSD'])
```

```python
tuples = list(zip(timestamp, id, token0, volumetoken0, token1, volumetoken1, volumeUSD))
```

```python

df = pd.DataFrame(tuples, columns=['timestamp', 'pairID', 'token0', 'volumetoken0', 'token1', 'volumetoken1', 'volumeUSD'])
```
